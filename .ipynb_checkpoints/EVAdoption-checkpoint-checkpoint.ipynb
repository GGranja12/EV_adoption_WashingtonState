```python
url = "https://electric-vehicle-data2.p.rapidapi.com/records"

states = ['AK', 'AL', 'AP', 'AR', 'AZ', 'BC', 'CA', 'CO', 'CT', 'DC','DE', 'FL', 'GA', 
            'HI', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA','MA', 'MD', 'MN', 'MO', 'MS', 'MT', 'NC', 'NE', 'NH', 'NJ',
            'NV', 'NY', 'OH', 'OR', 'PA', 'SC', 'TX', 'UT', 'VA','WA', 'WY']

headers = {
	"x-rapidapi-key": "017c936a82msh8860646249c65a4p13c03bjsn9438dd108cbd",
	"x-rapidapi-host": "electric-vehicle-data2.p.rapidapi.com"}

list_1 = []
for state in states:
    querystring = {"state":state,"orderBy":"dataListIndex_asc"}
    response = requests.get(url, headers=headers, params=querystring)
    data = response.json()
    data_1 = data["data"]
    list_1.extend(data_1)
list_1
```

```python
#Most relevant command!!


url = "https://electric-vehicle-data2.p.rapidapi.com/records"

states = ['AK', 'AL', 'AP', 'AR', 'AZ', 'BC', 'CA', 'CO', 'CT', 'DC','DE', 'FL', 'GA', 
          'HI', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA','MA', 'MD', 'MN', 'MO', 'MS', 'MT', 'NC', 'NE', 'NH', 'NJ',
          'NV', 'NY', 'OH', 'OR', 'PA', 'SC', 'TX', 'UT', 'VA','WA', 'WY']

headers = {
	"x-rapidapi-key": "017c936a82msh8860646249c65a4p13c03bjsn9438dd108cbd",
	"x-rapidapi-host": "electric-vehicle-data2.p.rapidapi.com"}

list_states = []

for state in states:
    for i in range (0, 5):
        querystring = {"state": state, "dataListIndex": f">={i*500}", "orderBy":"dataListIndex_asc"}
        response = requests.get(url, headers=headers, params=querystring)
        data = response.json()
        data_final = data["data"]
        list_states.extend(data_final)
```

```python
df_final = pd.DataFrame(list_states)
df_final.shape
```

```python
df_final.state.value_counts()
```
